```python
fig = plt.subplots(figsize=(15, 10)) 
ax = sns.violinplot(x="Country", y="LEABY", data=df)
plt.show()
```

Creating a point plot of GDP

```python
from matplotlib.ticker import FuncFormatter
def trillions(x, pos):
    'The two args are the value and tick position'
    return '$%1.1fT' % (x*1e-12)

formatter = FuncFormatter(trillions)


f, ax = plt.subplots(figsize=(15, 6)) 
ax = sns.pointplot(x="Year", y="GDP", hue = "Country", data=df)
plt.xticks(rotation=70)
ax.yaxis.set_major_formatter(formatter)
ax.set(ylabel="GDP in Trillions of U.S. Dollars")
plt.show()
```

Point plot of GDP over time

```python
from matplotlib.ticker import FuncFormatter
def trillions(x, pos):
    'The two args are the value and tick position'
    return '$%1.1fT' % (x*1e-12)

fig, ax = plt.subplots(figsize=(15, 10))
formatter = FuncFormatter(trillions)

sns.set_style("dark")
sns.set_palette("Dark2")
ax1 = sns.pointplot(x="Year", y="LEABY", hue = "Country", data=df)
ax2 = sns.pointplot(x="Year", y="GDP", hue = "Country", data=df)
ax2.yaxis.set_major_formatter(formatter)
ax1.set(ylabel="Life expectancy at birth (years)")
ax2.set(ylabel="GDP in Trillions of U.S. Dollars")
ax2.set_title("GDP in Trillions of Dollars Over Time")
ax1.set_title("Life Expectancy In Years Over Time")
```

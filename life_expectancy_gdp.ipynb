```python
# Import modules
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns

"""
Data Sources:
GDP Source: [World Bank] (http://data.worldbank.org/indicator/NY.GDP.MKTP.CD)national accounts data, 
and OECD National Accounts data files
Life Expectancy Data Source: [World Health Organization] (http://apps.who.int/gho/data/node.main.688)
"""
# Import data
df = pd.read_csv('all_data.csv')
df.rename(columns={'Life expectancy at birth (years)': 'LEABY'}, inplace=True)


# Graph data as a scatterplot
g = sns.FacetGrid(df, col='Year', hue='Country', col_wrap=4, height=2)
g = (g.map(plt.scatter, 'GDP', 'LEABY', edgecolor='w').add_legend())

# Graph GDP over time (2000-2015)
g2 = sns.FacetGrid(df, col='Country', col_wrap=3, height=4)
g2 = (g2.map(plt.plot, 'Year', 'GDP', color='m').add_legend())

# Graph Life Expectancy By Age over time (2000-2015)
g3 = sns.FacetGrid(df, col='Country', col_wrap=3, height=4)
g3 = (g3.map(plt.plot, 'Year', 'LEABY').add_legend())

```
